## Fonction zip()

```python
animaux = ["poulain", "renard", "python"]
couleurs = ["vert", "bleu", "rose"]

for animal, couleur in zip(animaux, couleurs):
    print(f"Le {animal} est {couleur}")
    
for i, animal, couleur in zip(range(3), animaux, couleurs):
    print(f"L'animal {i} ({animal}) est {couleur}")

for elt in zip(range(len(couleurs)), animaux, couleurs):
    print(elt)
```

## Type d'objet: **set**

```python
list = [1, 3, 5, 56, 5 , 6]
set(list)
s = set(list)

s.add(78)
s

s.discard(78)
s
```

```python
ligne = "MGALTESQAALVKSSWEEFNANIPKHTHRFFILVLEIAPAAKD"
s = set(ligne)
ligne2 = "EVVGAKWSEELNSA"
s2 = set(ligne2)

#union = joins two sets together
s_union = s.union(s2)
print(len(s), len(s2),len(s_union))
#intersection = commun dans les deux sets
```

## Copy of the lists

```python
list1 = [1, 2, 3]
#list2 = [3, 4, 5]
list2 = list1
list2[1] = -2356
print(list1, list2)
#changes both lists ( list2 points at the og list,
#and when we change list2 it changes the list object towards which we pointed)
```

```python
import copy
list1 = [1, 2, 3]
list2 = copy.deepcopy(list1)
list2[1] = -2356
print(list1, list2)
```
