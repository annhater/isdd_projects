### Exercice 2
Distance entre deux atomes carbones alpha consécutifs de la barstar

Obj: calculer la distance entre carbones alpha consécutifs le long de la chaîne peptidique avec module NumPy et de découvrir une anomalie

#### 1. Extraction des coordonnées atomiquesLecture des coordonnées

```python
with open("1BTA.pdb", "r") as f_pdb, open("1BTA_CA.txt", "w") as f_CA:
    for ligne in f_pdb:
        if ligne.startswith("ATOM") and ligne[12:16].strip() == "CA":
            x = ligne[30:38]
            y = ligne[38:46]
            z = ligne[46:54]
            f_CA.write(f"{x} {y} {z} ")
```

#### 2. Lecture des coordonnées

```python
with open("1BTA_CA.txt", "r") as f_CA:
#une liste contenant toutes les coordonnées sous forme de floats avec les fonctions split() et float()
#Affichez à l'écran le nombre total de coordonnées
    for line in f_CA:
        list_coord = [float(i) for i in line.split()] 
print(len(list_coord))
```

#### 3. Construction de la matrice de coordonnées

```python
arr_coord = np.array(list_coord)
#print(arr_coord)
print(arr_coord.reshape(89,3))
```

#### 4. Calcul de la distance

Créez maintenant une matrice qui contient les coordonnées des n-1 premiers carbones alpha et une autre qui contient les coordonnées des n-1 derniers carbones alpha. Affichez les dimensions des matrices pour vérification.
En utilisant les opérateurs mathématiques habituels (-, +, **2) et les fonctions sqrt() et sum() du module NumPy, calculez la distance entre les atomes n et n+1.

```python
n = 89
arr_coord = np.resize(arr_coord,(n,3))
print(arr_coord)
dist_mat_first = np.resize(arr_coord,(n-1,3))
print(dist_mat_first)
dist_mat_last = np.resize(arr_coord[1:],(n-1,3))
print(dist_mat_last)
```

```python
diff = dist_mat_first - dist_mat_last
distances = np.sqrt(np.sum(diff**2, axis=1))

for i, dist in enumerate(distances):
    print(f"Atoms {i+1}-{i+2}: {dist:.1f} Å")
    if dist > 4.5 or dist < 3.0:
        print(f"Valeur surprenante:{i+1} Å")
```
