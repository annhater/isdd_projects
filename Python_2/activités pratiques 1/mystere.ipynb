```python
#entre 1000 et 6000
#unités = u
#dizaines = d
#centaines = c
#milliers = m
#u=2c=3d
#m+c = 8

for i in range(1000, 6000+1):
    #make a str to iterate
    number = str(i)
    m = int(number[0])
    c = int(number[1])
    d = int(number[2])
    u = int(number[3])
    #check: print(number, m, c, d, u)
    #check the conditions
    if (u == 2*c) and (u == 3*d) and (m+c == 8):
        print(f"Le nombre mystère est : {number}")
```
