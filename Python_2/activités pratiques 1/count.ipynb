```python
def count_atoms(pdb_file):
    atoms_dico = {}
    #print(pdb_file) to check filename
    with open(pdb_file,"r") as filin:
        for line in filin:
            atom = line[76:78].strip()
            if atom not in atoms_dico:
                atoms_dico[atom] = 1
            else:
                atoms_dico[atom] += 1
    for key, value in atoms_dico.items():
        print(f"Atom {key} is found {value} times.")        
    most_common_atom = max(atoms_dico, key = atoms_dico.get)
    print(f"The most common atom is: {most_common_atom}.")

PDBNAME = "9ic4.pdb"
count_atoms(PDBNAME)
```

```python
def count_atoms(pdb_file):
    atoms_dico = {}
    #print(pdb_file) to check filename
    with open(pdb_file,"r") as filin:
        for line in filin:
            atom = line[76:78].strip()
            atoms_dico[atom] = atoms_dico.get(atom, 0) + 1
    for key, value in atoms_dico.items():
        print(f"Atom {key} is found {value} times.")        
    most_common_atom = max(atoms_dico, key = atoms_dico.get)
    print(f"The most common atom is: {most_common_atom}.")

PDBNAME = "9ic4.pdb"
count_atoms(PDBNAME)
```
