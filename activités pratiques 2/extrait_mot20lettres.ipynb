```python
liste_mots = []

with open("/usr/share/dict/french", "r") as dictionnaire_fr:
    for line in dictionnaire_fr:
        mot = line.strip()
        if len(mot) == 20:
            liste_mots.append(mot)
print(f"Le script a extrain {len(liste_mots)} mots de 20 lettres")

with open("mots_20lettres.txt", "w") as mots_20_out:
    for mots in liste_mots:
        mots_20_out.write(f"{mots}\n")
```
