### But: calculer et représenter l'axe de la première hélice alpha de la protéine 1bta.

### 1. Import libraries

```python
import numpy as np
```

### 2. Charger le fichier

```python
coords = np.loadtxt("helix1.dat")
print(coords)
coords.shape
```

### 3. Centrer les donnees


```python
coords_centered = (coords - coords.mean())/coords.std()
print(coords.mean(), coords.std())
print(coords_centered.mean(), coords_centered.std())
#coords_centered
```

```python
cov_mat = np.cov(coords_centered.T)
cov_mat.shape
cov_mat
#np.info(np.cov)
```

```python
#eigenvectors/eigenvalues

eigvals, eigvect = np.linalg.eig(cov_mat)
print(eigvals)
print(eigvect)
```

```python
axe = eigvect[:,0]
axe
```
