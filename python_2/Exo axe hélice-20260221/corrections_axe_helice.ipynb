```python
import numpy as np
```

```python
coords = np.loadtxt("helix1.dat")
coords
```

```python
coords_centrees = (coords - coords.mean()) / coords.std()
print(coords.mean(), coords.std())
```

```python
covmat = np.cov(coords_centrees.T)
covmat
```

```python
eigvals, eigvecs = np.linalg.eig(covmat)
print(eigvals)
eigvecs
```

```python
vec1 = eigvecs[:,0]
vec1
```

```python
print(coords)
# Centre de masse
point_depart = coords.mean(axis=0)
point_depart
```

```python
facteur = 10
point_arrivee = point_depart + vec1*facteur
point_arrivee
```
