## Dist Hamming

```python
def dist_Hamming(seq1, seq2):
    #for i in range(len(seq1)):
    #    print(seq1[i], seq2[i])
    dist = 0
    for aa1, aa2 in zip(seq1, seq2):
        if aa1 != aa2:
            dist += 1
    return dist
```

```python
dist_Hamming("AGWPSGGASAGLAIL", "IGWPSAGASAGLWIL")
```

```python
dist_Hamming("ATTCATACGTTACGATT", "ATACTTACGTAACCATT")
```

## Distance entre deux atomes carbones alpha consécutifs de la barstar

```python
with open("1bta.pdb", "r") as f_pdb, open("1bta_CA.txt", "w") as f_CA:
    for ligne in f_pdb:
        if ligne.startswith("ATOM") and ligne[12:16].strip() == "CA":
            x = ligne[30:38]
            y = ligne[38:46]
            z = ligne[46:54]
            f_CA.write(f"{x} {y} {z} ")
```

```python
with open("1bta_CA.txt", "r") as f:
    coords = [float(nb) for nb in f.read().split()]

print(f"Le nombre total de coordonnées est {len(coords)}")
```

```python
import numpy as np

coords_array = np.array(coords).reshape((89, 3))
print(coords_array.shape)
```

```python
coords_premiers = coords_array[:88,:]
print(coords_premiers.shape)
#print(coords_premiers)
```

```python
coords_derniers = coords_array[1:,:]
print(coords_derniers.shape)
#print(coords_derniers)
```

```python
diff = coords_derniers - coords_premiers
diff.shape
#diff
```

```python
carres_diffs = diff**2
carres_diffs.shape
```

```python
dist_carrees = carres_diffs.sum(axis=1)
print(dist_carrees.shape)
dist_carrees
```

```python
dists = np.sqrt(dist_carrees)
dists.shape
dists
```

```python
for i, dist in enumerate(dists):
    print(f"La distance entre les résidus {i+1} et {i+2} est {dist:.2f} angströms")
```

Pour trouver la valeur aberrante on peut utiliser un masque booléen

```python
dists[dists < 3.77]
```
