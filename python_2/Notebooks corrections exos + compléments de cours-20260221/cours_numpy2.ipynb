# Multiplication de matrice et transposée

```python
import numpy as np
a = np.arange(10).reshape((5, 2))
a
```

```python
a.T
```

```python
np.transpose(a)
```

```python
a
```

```python
np.dot(a, a.T)
```

Numpy recommande l'utilisation de l'opérateur `@` pour la multiplication de matrice. Notez que cela donne le même résultat que la fonction `np.dot()`.

```python
a @ a.T
```

```python
v1 = np.arange(5)
v1
```

```python
v2 = np.arange(10, 15)
v2
```

```python
v1 @ v2
```

Le produit matriciel de 2 vecteurs donne un scalaire, il s'agit en fait d'un produit scalaire.

# Matrice de Variance / Covariance

(parfois appelée matrice de covariance tout court)

```python
import random as rd
a = np.array([rd.randint(0, 10) for i in range(30)]).reshape((10, 3))
a
```

```python
# Centrer les données.
a = a - a.mean()
a.mean()
```

```python
# Standardiser les données.
a = a / a.std()
a.std()
```

```python
a
```

```python
# Attention on doit envoyer dans la fonction la transposée.
covmat = np.cov(a.T)
covmat
```

```python
eigvals, eigvecs = np.linalg.eig(covmat)
eigvals
```

# Tri

```python
a = np.array([rd.randint(0, 10) for i in range(10)])
a
```

```python
# Renvoie les indices triés par ordre croissant.
indices_tries = a.argsort()
indices_tries
```

```python
# On peut utiliser les indices triés pour l'indiçage de l'array, auquel cas on récupère l'array trié !
a[indices_tries]
```

```python
# On peut bien sûr récupérer l'array trié.
a = a[indices_tries]
a
```

# Matplotlib

(cf le [cours](https://python.sdv.u-paris.fr/21_module_matplotlib/) pour + de détails)

```python
import matplotlib.pyplot as plt

temps = [1, 2, 3, 4, 6, 7, 9]
concentration = [5.5, 7.2, 11.8, 13.6, 19.1, 21.7, 29.4]

fig, ax = plt.subplots()
ax.scatter(temps, concentration, marker="o", color="blue")
ax.set_xlabel("Temps (h)")
ax.set_ylabel("Concentration (mg/L)")
ax.set_title("Concentration de produit en fonction du temps")
```
