```python
def read_sequences(seq_file):
    seq_list = []
    with open(seq_file,"r") as filin:
        for seq in filin:
            seq_clean = seq.strip()
            seq_list.append(seq_clean)
        #print(seq_list)
    return seq_list


def compute_gc(sequence):
    prop_gc = (sequence.count("G")+sequence.count("C"))/len(sequence)
    return prop_gc

seq_file = "seq.txt"
sequences = read_sequences(seq_file)
#for seq_count, sequence in enumerate(sequences, start = 1):
#    print(f"Prop. GC sequence {seq_count}: {compute_gc(sequence):.2f}")
for seq_count, sequence in enumerate(sequences):
    print(f"Prop. GC sequence {seq_count+1}: {compute_gc(sequence):.2f}")
    
```
