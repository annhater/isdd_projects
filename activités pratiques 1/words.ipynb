```python
#import sys

#filename = sys.argv[1]

#if len.argv != 2:
#    sys.exit("ERROR: Please input only one argument(.txt filename)")

filename = "malediction_du_corbeau.txt"
malediction_dico = {}
word_counter = 0
with open (filename,"r") as filin:
    malediction_str = filin.read()
    malediction_clean = malediction_str.replace(".", "").replace(",","")
  
for word in malediction_clean:
    if word in malediction_dico:
        word_counter += 1
    else:
        malediction_dico.get(word,word_counter)
print(malediction_dico.items())


#print(f"Number of different words: {}")
#print(f"Longest word: {}")
#print(f"Shortest word: {}")
```
